```python
scaler = MinMaxScaler()
df[["HARGA RUMAH", "LT", "LB", "JKT", "JKM"]] = scaler.fit_transform(df[["HARGA RUMAH", "LT", "LB", "JKT", "JKM"]])
```

```python
encoder = LabelEncoder()
df["GRS"] = encoder.fit_transform(df["GRS"])
df["KOTA"] = encoder.fit_transform(df["KOTA"])
```
